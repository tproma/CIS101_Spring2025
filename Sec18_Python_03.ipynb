```python
#5:

a = int(input("a:"))
b = int(input("b:"))
c = input("Operation:")

if c == "+":
  print(a+b)
elif c == "-":
  print(a-b)
elif c == "*":
  print(a*b)
elif c == "/":
  print(a/b)
else:
  print("Invalid Operation.")
```
